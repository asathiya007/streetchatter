# StreetChatter

#### StreetChatter is an AI that assumes the role of Street Fighter characters (e.g. Ryu, Dhalsim, Chun-Li, M. Bison, Ken) and chats with users to help them learn the lore of that character and the Street Fighter universe. StreetChatter is powered by an LLM (Llama-3.3-70b-instruct) and retrieval augmented generation (RAG).

```python
from streetchatter import StreetChatter
```

```python
# utility function to display interaction between user and Street Fighter
# character
def display_interaction(sc, input):
    print(f'User: {input}\n')
    output = sc.invoke(input)
    print(f'\n{sc.character_name}: {output}')
```

## A conversation with Ryu

```python
ryu_sc = StreetChatter(character_name='Ryu')
```

```python
display_interaction(ryu_sc, 'Hi Ryu, tell me a bit about yourself!')
```

```python
display_interaction(ryu_sc, 'Cool, can you tell me about Ken and Sakura? How '
                    + 'have they helped you throughout your journey as a '
                    + 'fighter?')
```

```python
display_interaction(ryu_sc, 'What is the Satsui no Hado?')
```

```python
display_interaction(ryu_sc, 'Do you, Ken, Sakura, and Akuma all have the same '
                    + 'fighting style?')
```

```python
display_interaction(ryu_sc, 'I see, so would it be fair to say that you all '
                    + 'practice some form of Ansatsuken?')
```

```python
display_interaction(
    ryu_sc, 'Does the Satsui no Hado affect all practitioners of Ansatsuken?')
```

```python
display_interaction(
    ryu_sc, 'What are some of the moves you use in your fights?')
```

```python
display_interaction(ryu_sc, 'What is your strongest move?')
```

```python
display_interaction(
    ryu_sc, 'Cool, can you tell me about some of your most iconic battles?')
```

```python
display_interaction(
    ryu_sc, 'Can you tell me more about your battle with Sagat? How did you '
    + 'defeat him?')
```

```python
display_interaction(
    ryu_sc, 'Were you ever able to defeat Akuma? What about M. Bison?')
```

```python
display_interaction(ryu_sc, 'What is the Power of Nothingness?')
```

```python
display_interaction(
    ryu_sc, 'Thanks for the chat! I learned a lot. Bye!')
```
